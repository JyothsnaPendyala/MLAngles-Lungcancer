```python
# Setting target

X = df.drop('level', axis=1)
y = df.level
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
# Models
from sklearn import svm
from sklearn import tree

from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

import pickle

models = [svm.SVC(), tree.DecisionTreeClassifier()]
model_scores = []
for model in models:
    train = model
    fitting = train.fit(X_train, y_train)
    score = train.score(X_train, y_train)
    model_scores.append(score)
    y_predict2 = train.predict(X_test)
    
    print(model, ": ", score * 100,"%")
    print("\n\n\tClassification report")
    print(classification_report(y_test, y_predict2))
    print("\n\n\n")

    
print(model_scores)

best_model = pd.Series(model_scores).idxmax()
print("Best Model: ", models[best_model])

# saving the file
pickle.dump(models[best_model], open('best_model.pkl', 'wb'))
```

```python
# Loding the pickle file
pickle_model = pickle.load(open('best_model.pkl', 'rb'))
pickle_model.predict(X_test)
```
